```python
import pandas as pd
import numpy as np
import pyopenms as oms
from urllib.request import urlretrieve
from Bio import SeqIO
```

```python
DIR = "/path/to/data/"
```

```python
df_ep = pd.read_csv(DIR + "PDC000111_3.sdrf_openms_design_msstats_in.csv")
df_ep.head(5)
```

```python
df_ep2 = df_ep[["Reference", "ProteinName", "Intensity"]]
df_ep2
```

```python
df_ep2_gp = df_ep2.groupby(["Reference", "ProteinName"])["Intensity"].sum().reset_index()
df_ep2_gp
```

```python
#####################################
#####################################
```

```python
## prepare theoretical peptides
```

```python
list_protein = []
for i in SeqIO.parse("Homo-sapiens-uniprot-reviewed-entrap-contaminants-20240615.fasta","fasta"):
    j = [i.id, i.name, i.description, str(i.seq)]
    list_protein.append(j)
    #print(i.id, i.name, i.description, i.seq)
```

```python
list_protein
```

```python
df_seq = pd.DataFrame(list_protein, columns=["ID", "Name", "Description", "seq"])
df_seq
```

```python
### GET PEPTIDE

def get_peptide(bsa):
    dig = oms.ProteaseDigestion()
    #dig.getEnzymeName()  # Trypsin
    result =[]
    bsa = oms.AASequence.fromString(bsa)
    dig.digest(bsa, result, 7, 40)            ## peptide range 7, 40 aminoacid long

    return result

```

```python
df_seq["Trypsin"] = df_seq["seq"].apply(lambda x: get_peptide(x))
df_seq["Trypsin count"] = df_seq["Trypsin"].apply(lambda x: len(x))
df_seq
```

```python
df_seq2 = df_seq[["ID", "Trypsin count"]]
df_seq2
```

```python
# export to csv file
# df_seq2.to_csv("dict_trypsin.csv", index=False, sep="\t")
```

```python
# convert dataframe to dictionary
dict_trypsin = dict(zip(df_seq2["ID"], df_seq2["Trypsin count"]))
dict_trypsin
```

```python
## split ProteinName to list
df_ep2_gp["L_ProteinName"] = df_ep2_gp["ProteinName"].str.split(";")
df_ep2_gp
```

```python
df_ep2_gp4 = df_ep2_gp.explode("L_ProteinName")
df_ep2_gp4
```

```python
## map Trypsin count by L_ProteinName column
df_ep2_gp4["Trypsin count"] = df_ep2_gp4["L_ProteinName"].apply(lambda x: dict_trypsin[x])
df_ep2_gp4
```

```python
# calculate iBAQ
# iBAQ = Σintensity / #theoretical peptides
df_ep2_gp4["iBAQ"] = df_ep2_gp4["Intensity"] / df_ep2_gp4["Trypsin count"]
df_ep2_gp4
```

```python
## transform to abundance table
square_table = df_ep2_gp4.pivot(index="Reference", columns="L_ProteinName", values="iBAQ")
square_table
```

```python
square_table = square_table.fillna(0.00)
square_table
```

```python
## another way by removing row contain multiple ProteinName
df_ep2_gpa = df_ep2_gp[~df_ep2_gp["ProteinName"].str.contains(";")]
df_ep2_gpa
```

```python
df_ep2_gpa["Trypsin count"] = df_ep2_gpa["ProteinName"].apply(lambda x: dict_trypsin[x])
```

```python
df_ep2_gpa
```

```python
df_ep2_gpa["iBAQ"] = df_ep2_gpa["Intensity"] / df_ep2_gpa["Trypsin count"]
df_ep2_gpa
```

```python
square_table_a = df_ep2_gpa.pivot(index="Reference", columns="ProteinName", values="iBAQ")
square_table_a
```

```python
square_table_a = square_table_a.fillna(0)
square_table_a
```
