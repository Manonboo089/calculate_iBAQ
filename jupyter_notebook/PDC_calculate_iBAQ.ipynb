```python
import pandas as pd
import numpy as np
import pyopenms as oms
from urllib.request import urlretrieve
from Bio import SeqIO
```

```python
DIR = "/path/to/data/"
```

```python
df_trypsin = pd.read_csv(DIR + "uniprot_human_reference_isoforms_20180717_Trypsin_count.csv", sep="\t")
dict_trypsin = dict(zip(df_trypsin["ID"], df_trypsin["Trypsin count"]))
#dict_trypsin

df_lycs = pd.read_csv(DIR + "uniprot_human_reference_isoforms_20180717_LysC_count.csv", sep="\t")
dict_lycs = dict(zip(df_lycs["ID"], df_lycs["Lys-C count"]))
#dict_lycs
```

```python
df_pdc = pd.read_csv(DIR + "PDC000111_3.sdrf_openms_design_msstats_in.csv")
df_pdc.head(5)
```

```python
df_pdc2 = df_pdc[["Reference", "ProteinName", "Intensity"]]
df_pdc2["Sample_name"]   = df_pdc2["Reference"].str.split("_").str[:-2].str.join("_")
df_pdc2
```

```python
#### GROUP BY ["Sample_name", "ProteinName"] sum intensity
df_pdc2_gp = df_pdc2.groupby(["Sample_name", "ProteinName"], as_index=False).sum()
df_pdc2_gp
```

```python
df_pdc2_gp["L_ProteinName"] = np.where(df_pdc2_gp["ProteinName"].str.contains(";"), df_pdc2_gp["ProteinName"].str.split(";").str[0], df_pdc2_gp["ProteinName"])
df_pdc2_gp
```

```python
## mapping to trypsin count
## calculate iBAQ
# iBAQ = Σintensity / #theoretical peptides

df_pdc2_gp["Trypsin count"] = df_pdc2_gp["L_ProteinName"].apply(lambda x: dict_trypsin[x])
df_pdc2_gp["iBAQ"] = df_pdc2_gp["Intensity"] / df_pdc2_gp["Trypsin count"]
df_pdc2_gp
```

```python
square_table = df_pdc2_gp.pivot(index="Sample_name", columns="ProteinName", values="iBAQ")
square_table = square_table.fillna(0.00).round(2)
square_table
```

```python
square_table_T = square_table.T.reset_index()
square_table_T
```
